```python
import matplotlib.pyplot as plt
import numpy as np
df
```

```python
plt.subplots(figsize=(10,8))

df['gender'].value_counts(normalize=True)
df['gender'].value_counts(dropna=True).plot.bar(color='cyan')
plt.title('Comp. of male and female')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df['race/ethnicity'].value_counts()
```

```python
plt.subplots(figsize=(15,5))
df['race/ethnicity'].value_counts(normalize=True)
df['race/ethnicity'].value_counts(dropna=False).plot.bar(color=['black','red','blue','cyan'])
plt.title('Comp. of various groups')
plt.xlabel('Groups')
plt.ylabel('count')
plt.show()
```

```python
plt.subplots(figsize=(15,5))

df['math score'].value_counts(normalize=True)
df['math score'].value_counts(dropna=True).plot.bar(color='cyan')
plt.title('Comp. of math score')
plt.xlabel('Maths Score')
plt.ylabel('Count')
plt.show()
```

```python
plt.subplots(figsize=(10,8))

df['lunch'].value_counts(normalize=True)
df['lunch'].value_counts(dropna=True).plot.bar(color='yellow')
plt.title('Comp. of diff. types of lunch')
plt.xlabel('types of lunch')
plt.ylabel('Count')
plt.show()
```

```python
plt.subplots(figsize=(10,8))

df['test preparation course'].value_counts(normalize=True)
df['test preparation course'].value_counts(dropna=True).plot.bar(color=['green','red'])
plt.title('Comp. of diff. types of test preparation test completed or not')
plt.xlabel('Test completed or not')
plt.ylabel('Count')
plt.show()
```

```python
plt.subplots(figsize=(10,8))

df['reading score'].value_counts(normalize=True)
df['reading score'].value_counts(dropna=True).plot.bar(color=['orange'])
plt.title('Comp. of reading scores')
plt.xlabel('Reading score')
plt.ylabel('Count')
plt.show()
```

```python
plt.subplots(figsize=(10,8))

df['writing score'].value_counts(normalize=True)
df['writing score'].value_counts(dropna=True).plot.bar(color=['orange'])
plt.title('Comp. of writing scores')
plt.xlabel('Writing score')
plt.ylabel('Count')
plt.show()
```
